### Figure

グラフを描画する領域、少なくとも1以上のAxesが含まれる

Figureオブジェクトの作成

```python
fig = plt.figure()
```

### Axes(Subplot)

* 実際にグラフが描画されるオブジェクト
* Figureオブジェクトに属している
* Figureに複数のAxesを持てる

Axesオブジェクトの作成

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
```

#### add_subplotの引数

* 第1引数: 行数
* 第2引数: 列数
* 第3引数: Axesの順番

複数のAxesを作成

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)
```

FigureオブジェクトとAxesオブジェクトを同時に作成する場合

```python
fig, ax = plt.subplots()
```

## 3-2-2-代表的なグラフ

Axesオブジェクトからプロットするためのメソッドを呼び出す

|グラフの種類|メソッド|Y値 |X値   |
|------------|--------|----|------|
|線グラフ    |plot    |必須|省略可|
|棒グラフ    |bar     |必須|必須  |
|ヒストグラム|hist    |必須|不要  |
|箱ひげ図    |box     |必須|不要  |
|散布図      |scatter |必須|必須  |

各メソッドの引数にX値、Y値を渡す

### 線グラフ

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot([1, 2, 3], [5, 3, 7])
```

引数が1つの場合はX値に0からの自然数が割り振られる

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot([5, 3, 7])
```

### 棒グラフ

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.bar([1, 2, 3], [5, 3, 7])
```

X軸にラベルをつける

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.bar([1, 2, 3], [5, 3, 7], tick_label=["A", "B", "C"])
```

### 箱ひげ図

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.boxplot([1, 1, 2, 3, 3, 3, 3, 3, 3, 4, 4, 5, 6, 6, 6, 6, 7, 7, 7, 8, 8, 9, 10]);
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.boxplot([[1, 1, 2, 3, 3, 3, 3, 3, 3, 4, 4, 5, 6, 6, 6, 6, 7, 7, 7, 8, 8, 9, 10],
           [1, 2, 2, 3, 3, 3, 4, 4, 5, 6, 6, 7, 7, 8, 9, 12, 12, 13, 14, 15, 15, 16]]);
```

### ヒストグラム

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.hist([1, 1, 2, 3, 3, 3, 3, 3, 3, 4, 4, 5, 6, 6, 6, 6, 7, 7, 7, 8, 8, 9, 10]);
```

### 散布図

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.scatter([1, 2, 3], [5, 3, 7])
```

### 複数のグラフ

Axesオブジェクトに重ねて描画

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot([1, 2, 3], [5, 3, 7])
ax.bar([1, 2, 3], [2, 1, 3])
```

複数のAxesオブジェクトに描画

```python
fig = plt.figure()
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
ax1.plot([1, 2, 3], [5, 3, 7])
ax2.bar([1, 2, 3], [2, 1, 3])
```

## 3-2-3-扱えるデータ型

* タプル
* リスト
* numpy.ndarray
* pandas.Series
* pandas.DataFrame

```python
import numpy as np
import pandas as pd
```

### タプル

引数が1つの場合はY値として処理され、X値は0から自動的に自然数が割り振られる

```python
fig, ax = plt.subplots()
ax.plot((1, 2))
```

### リスト

```python
fig, ax = plt.subplots()
ax.plot([[1, 3, 5], [2, 4, 6]])
```

### numpy.ndarray

```python
arr = np.array([1, 2])
fig, ax = plt.subplots()
ax.plot(arr)
```

### pandas.Series

```python
ser1 = pd.Series([1, 2])
fig, ax = plt.subplots()
ax.plot(ser1)
```

引数が1つの場合はindexがX値として処理される

```python
ser2 = pd.Series([1, 3, 5], index=[0, 2, 10])
fig, ax = plt.subplots()
ax.plot(ser2)
```

### pandas.DataFrame

```python
df = pd.DataFrame([[1, 3, 5], [2, 4, 6]])
df
```

```python
fig, ax = plt.subplots()
ax.plot(df)
```

時系列データ（下記のデータ型）の場合は `register_matplotlib_converters` を実行し、 `matplotlib.units.registry` の内容を更新する

* pandas.Timestamp
* pandas.Period
* numpy.datetime64
* datetime.datetime
* datetime.date
* datetime.time

```python
from pandas.plotting import register_matplotlib_converters

register_matplotlib_converters()

ix = pd.date_range("2019-01-01", "2019-02-01")
ts = pd.Series(range(len(ix)), index=ix)

fig, ax = plt.subplots()
ax.plot(ts)
ax.xaxis.set_tick_params(rotation=45)  # X軸ラベルを45度回転
```

## 3-2-4-グラフのカスタマイズ

### Figureオブジェクトのサイズ

```python
fig = plt.figure(figsize=(3, 3))
ax1 = fig.add_subplot(1, 1, 1)
```

### タイトル

```python
fig = plt.figure()
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

fig.suptitle("Main Title")
ax1.set_title("Sub Title")
ax2.set_title("Sub Title")
```

### 凡例

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot([1, 2, 3], [5, 3, 7], label="data1")
ax.plot([1, 2, 3], [4, 7, 3], label="data2")
ax.legend()
```

### 書式指定

色の指定方法: [Specifying Colors](https://matplotlib.org/users/colors.html)

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(
    [1, 2, 3],  # X値
    [5, 3, 7],  # Y値
    color="Green",  # 線の色
    alpha=0.3,  # 不透明度
    linestyle="--",  # 線種
    linewidth=8,  # 線の太さ
)

fig.suptitle("Main Title", fontsize=15, color=(0, 0, 0))  # Figureのタイトル
ax.set_title("Title", fontsize=10, color="#FF1493")  # Axesのタイトル

ax.set_xlabel("x value", fontsize=15, color="DarkGreen")  # X軸ラベル
ax.set_xticks([1.5, 2.5])  # X軸目盛

ax.set_ylabel("y value", fontsize=15, color="Plum")  # X軸ラベル
ax.set_ylim(4, 6)  # Y軸の範囲
ax.tick_params(axis="y", labelsize=15, labelcolor="MediumVioletRed")  # Y軸ラベル
```

### Figure

グラフを描画する領域、少なくとも1以上のAxesが含まれる

Figureオブジェクトの作成

```python
fig = plt.figure()
```

### Axes(Subplot)

* 実際にグラフが描画されるオブジェクト
* Figureオブジェクトに属している
* Figureに複数のAxesを持てる

axesオブジェクトの作成

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
```

#### add_subplotの引数

* 第1引数: 行数
* 第2引数: 列数
* 第3引数: Axesの順番

複数のAxesを作成

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)
```

## 3-2-2-代表的なグラフ

Axesオブジェクトからプロットするためのメソッドを呼び出す

|グラフの種類|メソッド|
|------------|--------|
|線グラフ    |plot    |
|棒グラフ    |bar     |
|ヒストグラム|hist    |
|散布図      |scatter |
|円グラフ    |pie     |

各メソッドの引数にX値、Y値を渡す

### 線グラフ

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot([1, 2, 3], [5, 3, 7])
```

### 棒グラフ

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.bar([1, 2, 3], [5, 3, 7])
```

### ヒストグラム

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.hist([1, 1, 2, 3, 3, 3, 4, 4]);
```

### 散布図

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.scatter([1, 2, 3], [5, 3, 7])
```

### 複数のグラフ

```python
fig = plt.figure()
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
ax1.plot([1, 2, 3], [5, 3, 7])
ax2.bar([1, 2, 3], [5, 3, 7])
```

# 気象庁のデータの整形

|列名|説明|
|---------|---------|
|mean_temp|平均気温(℃)|
|max_temp |最高気温(℃)|
|min_temp |最低気温(℃)|
|humidity |平均湿度(％)|
|relative_humidity|最小相対湿度(％)|
|pressure|平均現地気圧(hPa)|
|precipitation|降水量の合計(mm)|
|duration|日照時間(時間)|
|wind|平均風速(m/s)|

```python
import pandas as pd

df = pd.read_csv('../data/data-chiba-2018.csv', encoding='shift_jis', skiprows=3)
df.dropna(inplace=True)
df.rename({"年月日": "date"}, axis=1, inplace=True)
df["date"] = pd.to_datetime(df["date"])
df.set_index("date", inplace=True)
df.drop(df.columns[~df.columns.str.endswith(")")], axis=1, inplace=True)
df.columns = [
    "mean_temp",
    "max_temp",
    "min_temp",
    "humidity",
    "relative_humidity",
    "pressure",
    "precipitation",
    "duration",
    "wind",
]
df = df.astype(float)
```
